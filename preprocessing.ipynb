```python
import cv2
import numpy as np

img = cv2.imread("archive/brain_tumor_dataset/no/1 no.jpeg", 0)  # grayscale
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
enhanced = clahe.apply(img)
```

```python
_, thresh = cv2.threshold(enhanced, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
kernel = np.ones((5,5), np.uint8)

# Remove noise
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Fill gaps
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel)
```

```python
num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(closing)

largest_label = 1 + np.argmax(stats[1:, cv2.CC_STAT_AREA])
brain_mask = (labels == largest_label).astype("uint8") * 255
```

```python
result = cv2.bitwise_and(img, img, mask=brain_mask)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(brain_mask, cmap='gray')
plt.title("Mask")

plt.subplot(1,3,3)
plt.imshow(result, cmap='gray')
plt.title("Skull Stripped")

plt.show()
```

```python
print("Original mean:", img.mean())
print("Processed mean:", result.mean())
```

```python
removed_pixels = (brain_mask == 0).sum()
total_pixels = brain_mask.size

print("Background removed %:", (removed_pixels / total_pixels) * 100)
```

```python
plt.hist(img.flatten(), bins=50, alpha=0.5, label="Original")
plt.hist(result.flatten(), bins=50, alpha=0.5, label="Processed")
plt.legend()
plt.title("Intensity Distribution")
plt.show()
```
